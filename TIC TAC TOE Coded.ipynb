```python
import math
import random
board = [[1,2,3],[4,5,6],[7,8,9]]
def display_board(board):
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print("|   ", board[0][0],"   |   ", board[0][1],"   |   ",  board[0][2],"   |", sep = "")
    print("|       |       |       |")
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print("|   ", board[1][0],"   |   ", board[1][1],"   |   ",  board[1][2],"   |", sep = "")
    print("|       |       |       |")
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print("|   ", board[2][0],"   |   ", board[2][1],"   |   ",  board[2][2],"   |", sep = "")
    print("|       |       |       |")
    print("+-------+-------+-------+")

def enter_move(board):
    # The function accepts the board current status, asks the user about their move,
    user_move = int(input("Please enter a number from 1 to 9.... "))
    # checks the input and updates the board according to the user's decision.
    if 1<= user_move<= 9 and isinstance((board[math.floor((user_move-1)/3)][(user_move-1) % 3]), int):
        board[math.floor((user_move-1)/3)][(user_move-1) % 3] = 'O'
        return board
    else: 
        print("That spot is taken or does not exist, try again.")
        enter_move(board)
    
def make_list_of_free_fields(board):
    # The function browses the board and builds a list of all the free squares; 
    empty = []
    for i in range(3):
        for m in range(3):
            if isinstance(board[i][m], int):  empty.append((i, m)) 
    # the list consists of tuples, while each tuple is a pair of row and column numbers.
    return empty
    
def victory_for(board, sign):
    # The function analyzes the board status in order to check if 
    if board[0][0] == board[0][1] == board[0][2] == sign or board[1][0] == board[1][1] == board[1][2] == sign\
    or  board[2][0] == board[2][1] == board[2][2] == sign or board[0][0] == board[1][0] == board[2][0] == sign\
    or board[0][1] == board[1][1] == board[2][1] == sign or board[0][2] == board[1][2] == board[2][2] == sign\
    or board[0][0] == board[1][1] == board[2][2] == sign or board[0][2] == board[1][1] == board[2][0] == sign:
        print("The winner is... ", sign, "!!!!", sep = "")
        return True
    # the player using 'O's or 'X's has won the game

def check_draw(board):
    if victory_for(board, 'X')!= True and victory_for(board, 'O') != True and\
    all(isinstance(item, str) for item in board[0]) and all(isinstance(item, str) for item in board[1]) and\
    all(isinstance(item, str) for item in board[2]):
        print("The match is a draw!!!")
        return True
def draw_move(board):
    # The function draws the computer's move and updates the board.
    row, column = random.choice(make_list_of_free_fields(board))
    board[row][column] = 'X'

def tic_tac_toe(board):
    while True:
        draw_move(board)
        display_board(board)
        victory_for(board, 'X')
        victory_for(board, 'O')
        check_draw(board)
        if check_draw(board) or victory_for(board, 'X') or victory_for(board, 'O'): exit()
        enter_move(board)
        display_board(board)
        check_draw(board)
        victory_for(board, 'X')
        victory_for(board, 'O')
        if check_draw(board) or victory_for(board, 'X') or victory_for(board, 'O'): break
tic_tac_toe(board)
```
